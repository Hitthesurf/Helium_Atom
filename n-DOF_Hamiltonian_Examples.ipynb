## Change Sign Of V

We have matrices $A_1, A_2$ with eigenvalues $\lambda_1, \lambda_2$.

If $A_1 = -A_2$ then $\lambda_1 = -\lambda_2$

This is because
> \begin{align*}
det(A_2-\lambda_2I) &= 0\\
det(-A_1-\lambda_2I) &= 0\\
det(-(A_1+\lambda_2I)) &= 0\\
det(-I(A_1+\lambda_2I)) &= 0
\end{align*}

From the fact $det(AB)=det(A)det(B)$ and $det(-I)= \pm 1$

> \begin{align*}
det(-I(A_1+\lambda_2I)) &= 0\\
det(-I)det(A_1+\lambda_2I) &= 0\\
det(A_1+\lambda_2I) &= 0
\end{align*}

For this to be true $\lambda_1 = -\lambda_2$. This is important as we can determine what happens to the eigenvalues of $\frac{\partial V}{\partial q_i \partial q_j}$ matrix (Hessain of $V$). This means when V has a change of sign this causes the matrix to change its sign. Therefore the eigenvalues change their sign and the nature of the equilibrium points will invert (centre to saddle, saddle to centre).



## Stability of Periodic Orbit Using Monodromy Matrix

We have $y_{po}$ is a periodic orbit with period $T$, in phase space $\textbf{$x$}$. So $y_{po}(t)=y_{po}(t+T_n)$ for all $n \in \mathbb{Z}$.

Let $\delta$ be a small displacement of the periodic orbit at time.

The basic idea is

* Get position $y_{po}(0)$, step off the periodic orbit by small amount $y^{'}(0) = y_{po}(0) + \delta$ (Where $y^{'}$ is a new trajectory)
* Let the orbit $y^{'}$ flow for time $T$.
* $y^{'}(T) = y_{po}(0) + \delta^{'}$

By comparing points $y^{'}(T),y_{po}(0), y^{'}(0)$ we can deduce the nature of the periodic orbit.

We have that the monodromy matrix is the same as the fundamental solution matrix at time $T$.

So we can find the monodromy matrix by using these equations.
> \begin{align*}
\dot{M} &= LM\\
\leftrightarrow M(T) &= e^{LT}
\end{align*}


Where $L$ is a linearisation of the flow along the periodic trajectory. Therefore,

Let $\gamma = (q_1,q_2,...q_n,p_1,p_2,...,p_n)$

Let $J$ be the matrix 
$$
\left[\begin{array}{cc} 
0 & I\\
-I & 0
\end{array}\right]
$$ 

Let $F(x_1,x_2,...,x_n) = (\dot{x_1},\dot{x_2},...,\dot{x_n})$
> $L = J\frac{\partial^2 H}{\partial \gamma \partial \gamma} = \nabla(F)$

If the linearisation is time-dependent, it must be evaluated at the periodic orbit of interest. 

We then find the eigenvalues and eigenvectors of $M(T)$, $\lambda_i, v_i$

We have the fact, if we step off the periodic orbit in the direction of an eigenvector $v_i$, when we come back around after waiting time $T$ we will be step*$\lambda_i$ away from periodic orbit. 

So for a periodic orbit
* If $|\lambda_i|>1$ unstable in direction $v_i$
* If $|\lambda_i|<1$ asymptotically stable in direction $v_i$
* If $|\lambda_i|=1$ lyapunov stable in direction $v_i$

We also have, Hamiltonian System $\rightarrow det(M)=1$, also to find the $T$ of a periodic orbit in a phase plane we can do the lcm of all the periodic orbits in each plane.

The best way to find do this computation is to use wolfram alpha.

# Examples

## Non Hamiltonian Examples

### Example 1

> \begin{align*}
\dot{x} &= -y\\
\dot{y} &= x\\
\dot{z} &= -z
\end{align*}

Here we have an equilibrium at $(0,0,0)$ and periodic orbits in the $x,y$ plane with $x^2+y^2=r^2$.

We will fix $r$.

We only get stable manifolds which are.

A line $W^s((0,0,0)) = \{(x,y,z)\in \mathbb{R}^3: x=0, y=0\}$

A surface which is a cylinder
$W^s(\{x^2+y^2=r^2,z=0\}) = \{(x,y,z)\in \mathbb{R}^3: x^2 +y^2 = r^2, z \in \mathbb{R} \}$

#### Stability Of Periodic Orbit By Monodromy Matrix

We have, $F(x,y,z)=(-y,x,-z)$ and
$$
L=\nabla(F)=\left[\begin{array}{cc} 
0 & -1 & 0\\
1 & 0  & 0\\
0 & 0  & -1
\end{array}\right]
$$ 
Then find monodromy matrix at $t$
$$
M(t)=e^{Lt}=\left[\begin{array}{cc} 
cos(t) & -sin(t) & 0\\
sin(t) & cos(t)  & 0\\
0 & 0  & e^{-t}
\end{array}\right]
$$ 
Now find monodromy matrix at $T$
$$
M(T)=e^{LT}=\left[\begin{array}{cc} 
1 & 0 & 0\\
0 & 1  & 0\\
0 & 0  & e^{-T}
\end{array}\right]
$$ 
We have $det(M(T))$ is not equal to $1$ as the ODE system is not conservative.

We get eigenvalues and eigenvectors.
> \begin{align*}
\lambda_1 &= e^{-T},& v_1 =(0,0,1)\\
\lambda_2 &= 1,& v_2 =(0,1,0) \\
\lambda_3 &= 1,& v_3 =(1,0,0) 
\end{align*}

So in the z direction $v_1$, we have the periodic orbit is asymptotically stable.

And in the plane $(x,y)$, we have it is lyapunov stable.

Since all eigenvalues are $\leq 1$, then we have the periodic orbit is stable.


```python
def example1(t, x1, x2, x3):
    f1 = -1*x2
    f2 = 1*x1
    f3 = -x3
    return np.array([f1, f2, f3])

ODE = ODEAnalysis(example1)
ODE.PlotPhaseSpace3D([-2,2],[-2,2],[-2,2],5, gap = 2)
```

```python
# CalcEigenValues
def example1Monodromy(t, x1, x2, x3, M):
    L = np.matrix([[0, -1, 0],
                   [1, 0, 0],
                   [0, 0, -1]])

    f1 = -x2
    f2 = x1
    f3 = -x3
    
    M_dot = L*M
    
    return np.array([f1,f2,f3,M_dot])

ODE = ODEAnalysis(example1Monodromy, StepOfTime= 0.001)
ODE.ShowEigen(2*np.pi, [1,3,0], IntegrateOnOrbit = True)

ODE.getMonoDromy(2*np.pi, [1,3,0])
```

So our eigenvalues are correct, but are eigenvectors are slightly different this is because we did it numerically and this means the monodromy matrix calculated at time $T$ is slightly off so our eigenvectors will not be the exact ones.

## Hamiltonian Examples

### 1-Dof 

#### Example 2

> \begin{align*}
\dot{x} &= \lambda p_x\\
\dot{p_x} &= \lambda x
\end{align*}

By calculating the Hamiltonian we get 
$H = \frac{\lambda}{2}(p_x^2-x^2)$

This gives us the phase plane

```python
def example2(t, x1, x2):
    # lambda = 1
    # p_x is y
    
    f1 = 1*x2
    f2 = 1*x1
    return np.array([f1, f2])

ODE = ODEAnalysis(example2)
ODE.PlotPhasePlane([-4,4],[-3.5,3.5],5)
```

##### Equilibrium:
We have equilibrium at $(0,0)$

##### Nature:
We have $V = - \lambda x^2/2$. So the Hessian matrix of V is $[\lambda]$ in our case we have taken $\lambda > 0$ so we get a saddle at $(0,0)$ in the $x,p_x$ plane.

##### Trajectories:
Since this is Hamiltonian all the trajectories for a given $h$ have equation $h = \frac{\lambda}{2}(p_x^2-x^2)$.

##### Periodic Orbits:
The only periodic orbit is the equilibrium $(0,0)$ but this isn't really a periodic orbit.

##### Manifolds:
we have

$W^s((0,0)) = \{(x,p_x)\in\mathbb{R}^2:p_x = -x\}$

$W^u((0,0)) = \{(x,p_x)\in\mathbb{R}^2:p_x = x\}$

##### Stability Of Periodic Orbit By Monodromy Matrix
We have, $H = \frac{\lambda}{2}(p_x^2-x^2)$, $\gamma = (x,p_x)$ and
$$
L=J\frac{\partial^2 H}{\partial \gamma \partial \gamma}=\left[\begin{array}{cc} 
0 & \lambda\\
\lambda & 0
\end{array}\right]
$$ 
Then find monodromy matrix at $t$
$$
M(t)=e^{Lt}=\left[\begin{array}{cc} 
cosh(\lambda t) & sinh(\lambda t)\\
sinh(\lambda t) & cosh(\lambda t)
\end{array}\right]
$$ 
Now find monodromy matrix at $T$
$$
M(T)=e^{LT}=\left[\begin{array}{cc} 
cosh(\lambda T) & sinh(\lambda T)\\
sinh(\lambda T) & cosh(\lambda T)
\end{array}\right]
$$ 
We have $det(M(T)) = 1$ as the system is Hamiltonian.

We get eigenvalues and eigenvectors.
> \begin{align*}
\lambda_1 &= cosh(\lambda T) - sinh(\lambda T) = e^{-\lambda T},& v_1 =(-1,1)\\
\lambda_2 &= cosh(\lambda T) + sinh(\lambda T) = e^{\lambda T},& v_2 =(1,1)
\end{align*}

We have periodic orbit is asymptotically stable in direction $v_1$

We have periodic orbit is unstable in direction $v_1$

Since not all eigenvalues are $\leq 1$, then we have the periodic orbit is unstable.


```python
# CalcEigenValues
def example2Monodromy(t, q1, p1, M):
    a = 1
    L = np.matrix([[0, a],
                   [a, 0]])

    dot_q1 = a*p1
    dot_p1 = a*q1
    
    M_dot = L*M
    
    return np.array([dot_q1, dot_p1, M_dot])

ODE = ODEAnalysis(example2Monodromy, StepOfTime= 0.001)
ODE.ShowEigen(2*np.pi, [0,0], IntegrateOnOrbit = True)

ODE.getMonoDromy(2*np.pi, [1,3])
```

### 2-DOF 

#### Example 3

$\lambda , w > 0$

> \begin{align*}
\dot{x} &= \lambda p_x\\
\dot{p_x} &= \lambda x\\
\dot{y} &= wp_y\\
\dot{p_y} &= -wy
\end{align*}

This gives us $H = \frac{\lambda}{2}(p_x^2-x^2) + \frac{w}{2}(p_y^2+y^2)$

Since this is decoupled we can split this into two 1-DoF cases with
> \begin{align*}
H_1 &= \frac{\lambda}{2}(p_x^2-x^2)\\
H_2 &= \frac{w}{2}(p_y^2+y^2)
\end{align*}

```python
# H_1 system same as previous example
# Take H_2 system
def example3(t, x1, x2):
    # w = 1
    # x2 is p_y on the yaxis
    # x1 is y on the xaxis
    
    f1 = 1*x2
    f2 = -1*x1
    return np.array([f1, f2])

ODE = ODEAnalysis(example3)
ODE.PlotPhasePlane([-4,4],[-3.5,3.5],5)
```

##### Equilibrium:
We have equilibrium at  (0,0,0,0) 

##### Nature:
We have $V_1 = -\frac{\lambda x^2}{2}$ and $V_2 = -\frac{\lambda y^2}{2}$, so $V = V_1+V_2 = -\frac{\lambda x^2}{2} + \frac{\lambda y^2}{2}$. The Hessian of $V$ is

$$
\left[\begin{array}{cc} 
-\lambda & 0\\
0 & w
\end{array}\right]
$$ 

Therefore, at (0,0,0,0)

We have a saddle in the plane of $(x,p_x)$

We have a centre in the plane of $(y,p_y)$

##### Trajectories:
We have a fixed total energy $h$. Since we have an uncoupled system in the $(x,p_x)$ and $(y,p_y)$ plane, we can fix $h_1$ and $h_2$ such that $h = h_1 + h_2$

The trajectories in the plane $(x,p_x)$ given $h_1$ are $h_1= \frac{\lambda}{2}(p_x^2-x^2)$

The trajectories in the plane $(y,p_y)$ given $h_2$ are $h_2= \frac{w}{2}(p_y^2+y^2)$

##### Periodic Orbits:
We have non-trivial periodic orbits only in $(y, p_y)$ which are circles. With equation $h_2= \frac{w}{2}(p_y^2+y^2)$ thus the bigger the energy the bigger the circle/orbit. But for the trajectory to be a periodic orbit in the 4D phase space we must have a periodic orbit in the $(x, p_x)$ thus $(x, p_x)=(0,0)$ as equilibrium in plane $(x,p_x)$ acts as periodic orbit. This means $h_1 = 0, h = h_2$.

##### Unstable & Stable Manifolds:

$W^s(\{(x,p_x,y,p_y)\in \mathbb{R}^4: x=p_x=0, 2h_2=w(p_y^2+y^2)\}) = \{(x,p_x,y,p_y)\in \mathbb{R}^4:2h_2=w(p_y^2+y^2), p_x=-x\}$

$W^u(\{(x,p_x,y,p_y)\in \mathbb{R}^4: x=p_x=0, 2h_2=w(p_y^2+y^2)\}) = \{(x,p_x,y,p_y)\in \mathbb{R}^4:2h_2=w(p_y^2+y^2), p_x=x\}$

This makes sense as along $p_x = \pm x$, $h_1=0$, and $h_2$ remains constant so the manifolds are isoenergetic (constant energy) which they should be as they are trajectories.

##### Stability Of Periodic Orbit By Monodromy Matrix
We have, $F(x,y,p_x,p_y)=(\lambda p_x, w p_y, \lambda x, -wy)$ and
$$
L=L=\nabla(F)=\left[\begin{array}{cc} 
0 & 0 & \lambda & 0\\
0 & 0 & 0 & w\\
\lambda & 0 & 0 & 0\\
0 & -w & 0 & 0
\end{array}\right]
$$ 
Then find monodromy matrix at $t$
$$
M(t)=e^{Lt}=\left[\begin{array}{cc} 
cosh(\lambda t) & 0 & sinh(\lambda t) & 0\\
0 & cos(wt) & 0 & sin(wt)\\
sinh(\lambda t) & 0 & cosh(\lambda t) & 0\\
0 & -sin(wt) & 0 & cos(wt)
\end{array}\right]
$$ 
Now find monodromy matrix at $T$
$$
M(T)=e^{LT}=\left[\begin{array}{cc} 
cosh(\lambda T) & 0 & sinh(\lambda T) & 0\\
0 & 1 & 0 & 0\\
sinh(\lambda T) & 0 & cosh(\lambda T) & 0\\
0 & 0 & 0 & 1
\end{array}\right]
$$ 
We have $det(M(T)) = 1$ as the system is Hamiltonian.

We get eigenvalues and eigenvectors.
> \begin{align*}
\lambda_1 &= 1,& v_1 = (0,0,0,1)\\
\lambda_2 &= 1,& v_2 = (0,1,0,0)\\
\lambda_3 &= e^{-\lambda T},& v_3 = (-1,0,1,0)\\
\lambda_4 &= e^{\lambda T},& v_4 = (1,0,1,0)
\end{align*}

We have periodic orbit is lyapunov stable in plane $(y,p_y)$

We have periodic orbit is asymptotically stable in direction $v_3$

We have periodic orbit is unstable in direction $v_4$

Since not all eigenvalues are $\leq 1$, then we have the periodic orbit is unstable.

```python
# CalcEigenValues
def example3Monodromy(t, q1, q2, p1, p2, M):
    a = 1
    b = 1
    L = np.matrix([[0, 0, a, 0],
                   [0, 0, 0, b],
                   [a, 0, 0, 0],
                   [0, -b, 0, 0]])

    dot_q1 = a*p1
    dot_q2 = b*p2
    dot_p1 = a*q1
    dot_p2 = -b*q2

    M_dot = L*M

    return np.array([dot_q1, dot_q2, dot_p1, dot_p2, M_dot])


ODE = ODEAnalysis(example3Monodromy, StepOfTime=0.001)
ODE.ShowEigen(2*np.pi, [0, 3,0,3],IntegrateOnOrbit = True)

ODE.getMonoDromy(2*np.pi, [0, 3,0,3])
```

#### Example 4

> \begin{align*}
\dot{q}_1 &= p_1\\
\dot{p}_1 &= -q _1+q_1^3\\
\dot{q}_2 &= p_2\\
\dot{p}_2 &= -q _2
\end{align*}

This gives us $H = \frac{p_1^2}{2}+\frac{q_1^2}{2}-\frac{q_1^4}{4} + \frac{p_2^2}{2}+\frac{q_2^2}{2}$

Since this is decoupled we can split this into two 1-DoF cases with
> \begin{align*}
H_1 &= \frac{p_1^2}{2}+\frac{q_1^2}{2}-\frac{q_1^4}{4}\\
H_2 &= \frac{p_2^2}{2}+\frac{q_2^2}{2}
\end{align*}

```python
# H_2 system same as previous phase plane example
# Take H_1 system
def example4(t, x1, x2):
    # x2 is p_1 on the yaxis
    # x1 is q_1 on the xaxis
    
    f1 = 1*x2
    f2 = -1*x1+x1**3
    return np.array([f1, f2])

ODE = ODEAnalysis(example4)
ODE.PlotPhasePlane([-3,3],[-3,3],5, gap = 0.5)
```

##### Equilibrium
We have equilibrium at $(q_1,p_1,q_2,p_2) = (0,0,0,0),(0.\pm 1,0,0)$

##### Nature
We have $V_1 = \frac{q_1^2}{2}-\frac{q_1^4}{4}$ and $V_2 = \frac{q_2^2}{2}$

So $V = V_1 + V_2 = \frac{q_1^2}{2}-\frac{q_1^4}{4} + \frac{q_2^2}{2}$

The Hessian of V is
$$
\left[\begin{array}{cc} 
1-3q_1^2 & 0\\
0 & 1
\end{array}\right]
$$ 

This means we have eigenvalues $\lambda = 1-3q_1^3,1$

At the equilibrium $(0,0,0,0)$ the eigenvalues are $\lambda = 1,1$. So we have in the plane $(q_1,p_1)$ a centre, in the plane $(q_2,p_2)$ we also have a centre.

At $(q_1,p_1,q_2,p_2) = (0.\pm 1,0,0)$ the hessian of V has eigenvalues $\lambda = -2,1$. So in the plane $(q_1,p_1)$ we have a saddle, in the plane $(q_2,p_2)$ we have a centre.

##### Trajectories
We have a fixed total energy $h$. Since we have an uncoupled system in the $(q_1,p_1)$ and $(q_2,p_2)$ plane, we can fix $h_1$ and $h_2$ such that $h = h_1 + h_2$

The trajectories in the plane $(q_1,p_1)$ given $h_1$ are $h_1= \frac{p_1^2}{2}+\frac{q_1^2}{2}-\frac{q_1^4}{4}$

The trajectories in the plane $(q_2,p_2)$ given $h_2$ are $h_2= \frac{p_2^2}{2}+\frac{q_2^2}{2}$

##### Stable Periodic Orbits
We have stable periodic orbits in $(q_1,p_1)$ when 
> $0 \leq (h_1 = \frac{p_1^2}{2}+\frac{q_1^2}{2}-\frac{q_1^4}{4}) < \frac{1}{4}$

This can easily be calculated by subbing $(q_1, p_1) = (1,0),(0,0)$ into $H_1$ as we want the energy levels between the points $(1,0),(0,0)$. Since the trajectories form periodic orbits.

$h_1$  not $= \frac{1}{4}$ as this would not be a periodic orbit as the trajectory would contain an equilibrium point which it would never get through. It would be a heteroclinic orbit.

Also this equation gives multiple trajectories the ones we want are $|q_1|<1$

Additionally, we have all the orbits in the plane $(q_2,p_2)$ are periodic. All these periodic orbits are lyapunov stable.

This leads us to the set of stable periodic orbits

$\{(q_1,p_1,q_2,p_2) \in \mathbb{R}^4: 0 \leq (h_1 = \frac{p_1^2}{2}+\frac{q_1^2}{2}-\frac{q_1^4}{4}) < \frac{1}{4}, |q_1|<1, 2h_2= p_2^2+q_2^2\}$

##### Unstable Periodic Orbits
We also have unstable periodic orbits, this occurs for any energy $h_2 \geq 0$ as in the plane $(q_2,p_2)$ every orbit is periodic. But we must have $h_1 = \frac{1}{4}$ as this is the energy where the trajectories have equilibrium saddle points in the plane $(q_1,p_1)$ with $(q_1,p_1)=(\pm 1,0)$. 

Unstable periodic orbits are

$\{(q_1,p_1,q_2,p_2) \in \mathbb{R}^4: q_1 = \pm 1, p_1 = 0, 2h_2 = p_2^2+q_2^2\}$

##### Stable and Unstable Manifolds
Since the stable periodic orbits are lyapunov stable they don't have a manifold.

But the unstable periodic orbits do have manifolds.

Since it is a decoupled system we first calculate the stable/unstable manifolds of the phase plane $(q_1,p_1)$ at the equilibrium $(q_1,p_1) = (\pm 1, 0)$.

We have $h_1=1/4$. Therefore, the trajectories we are interested in are
> $\frac{1}{4} = \frac{p_1^2}{2} + \frac{q_1^2}{2} - \frac{q_1^4}{4} \leftrightarrow q_1^2 = 1 \pm p_1 \sqrt{2}$

It is clear from the phase plane plot of $(q_1,p_1)$
> \begin{align*}
W^s(\{(q_1,p_1)=(+1,0)\}) &= \{(q_1,p_1)\in \mathbb{R}^2: q_1^2 = 1-p_1\sqrt{2}, q_1>-1\}\\
W^u(\{(q_1,p_1)=(+1,0)\}) &= \{(q_1,p_1)\in \mathbb{R}^2: q_1^2 = 1+p_1\sqrt{2}, q_1>-1\}\\
W^s(\{(q_1,p_1)=(-1,0)\}) &= \{(q_1,p_1)\in \mathbb{R}^2: q_1^2 = 1+p_1\sqrt{2}, q_1<1\}\\
W^u(\{(q_1,p_1)=(-1,0)\}) &= \{(q_1,p_1)\in \mathbb{R}^2: q_1^2 = 1-p_1\sqrt{2}, q_1<1\}
\end{align*}

Then we need to calculate the manifolds for the phase space, we said unstable periodic orbits can have any energy in $h_2 \geq 0$. So we just put this into our definition from before.
> \begin{align*}

The role of the stable manifolds is the possible trajectories a particle can take so as $t \rightarrow \infty$ it will approach the periodic orbit.

#### Example 4 and 5 Using Monodromy Matrix To Find Nature Of Periodic Orbit

We have the parameter $\beta$.

> \begin{align*}
\dot{q}_1 &= p_1\\
\dot{p}_1 &= -\beta-q _1+q_1^3\\
\dot{q}_2 &= p_2\\
\dot{p}_2 &= -q _2
\end{align*}

This gives us $H = \frac{p_1^2}{2}+\frac{q_1^2}{2}-\frac{q_1^4}{4} +\beta q_1+ \frac{p_2^2}{2}+\frac{q_2^2}{2}$

We have, $\gamma = (q_1,q_2,p_1,p_2)$, $k^2=3q_1^2 -1$ and
$$
L=J\frac{\partial^2 H}{\partial \gamma \partial \gamma}=\left[\begin{array}{cc} 
0 & 0 & 1 & 0\\
0 & 0 & 0 & 1\\
k^2 & 0 & 0 & 0\\
0 & -1 & 0 & 0
\end{array}\right]
$$ 
Then find monodromy matrix at $t$
$$
M(t)=e^{Lt}=\left[\begin{array}{cc} 
cosh(kt) & 0 & \frac{1}{k}sinh(kt) & 0\\
0 & cos(t) & 0 & sin(t)\\
ksinh(kt) & 0 & cosh(kt) & 0\\
0 & -sin(t) & 0 & cos(t)
\end{array}\right]
$$ 
Now find monodromy matrix at $T$, Since periodic $q_1(T)=q_1(0)$, $k_0 =+ \sqrt{3q_1^2(0) - 1} = k(T)$ and we will take $p_1(T) = q_1(0) = 0$.

So in this case $T = lcm(2 \pi, \text{time period of periodic orbit starting at $q_1(0)$})$
$$
M(T)=e^{LT}=\left[\begin{array}{cc} 
cosh(k_0T) & 0 & \frac{1}{k_0}sinh(k_0T) & 0\\
0 & 1 & 0 & 0\\
k_0sinh(k_0T) & 0 & cosh(k_0T) & 0\\
0 & 0 & 0 & 1
\end{array}\right]
$$ 
We have $det(M(T)) = 1$ as the system is Hamiltonian.

We get eigenvalues and eigenvectors.
> \begin{align*}
\lambda_1 &= 1,& v_1 = (0,0,0,1)\\
\lambda_2 &= 1,& v_2 = (0,1,0,0)\\
\lambda_3 &= e^{-k_0T},& v_3 = (-\frac{1}{k_0},0,1,0)\\
\lambda_4 &= e^{k_0T},& v_4 = (\frac{1}{k_0},0,1,0)\\
\end{align*}

We have periodic orbit is lypunov stable in the plane $(q_2,p_2)$ for all $q_1$ 

We have periodic orbit is unstable when $q_1 > \frac{\sqrt{3}}{3}$in direction $v_4$, and is stable in direction $v_3$ when $q_1 > \frac{\sqrt{3}}{3}$

Therefore, when $q_1 > \frac{\sqrt{3}}{3}, p_1 = 0$ we have the periodic orbit is unstable.

When $q_1 < \frac{\sqrt{3}}{3}, p_1 = 0$ we have $|\lambda_3| = |\lambda_4| = 1$ as $k_0$ becomes imagery. Therefore, the periodic orbit is lypunov stable. 

This disagrees with our result earlier. This makes sense because if the periodic orbit was near the edge of the stable region we defined earlier, then a small displacement away form the origin would lead to it no longer being a periodic orbit. 


```python
# CalcEigenValues
def example4and5Monodromy(t, q1, q2, p1, p2, M):
    b=0
    L = np.matrix([[0, 0, 1, 0],
                   [0, 0, 0, 1],
                   [3*q1**2-1, 0, 0, 0],
                   [0, -1, 0, 0]])

    dot_q1 = p1
    dot_p1 = -b-q1+q1**3
    dot_q2 = p2
    dot_p2 = -q2

    M_dot = L*M

    return np.array([dot_q1, dot_q2, dot_p1,dot_p2, M_dot])


ODE = ODEAnalysis(example4and5Monodromy, StepOfTime=0.001)
```
