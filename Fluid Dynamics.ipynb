```python
from ODEAnalysis import *
import warnings
import numpy as np
```

```python
def example2(t, x1, x2):
    # lambda = 1
    # p_x is y
    
    f1 = -1*x2
    f2 = 1*x1 + t
    return np.array([f1, f2])

ODE = ODEAnalysis(example2)
ODE.PlotPhasePlane([-4,4],[-3.5,3.5],2, gap = 2)
```

```python
ODE = ODEAnalysis(example2)
ODE.StreakLine(0,3.14,[0,0], plot = True)
```

```python
ODE = ODEAnalysis(example2)
final_x = ODE.StreakLine(0,3.14,[0,0])
x = final_x[:,0]
y = final_x[:,1]

t = np.arange(0,3.14,0.01)
your_x = -(3.14-t)+np.sin(t)
your_y = 1 + np.cos(t)

my_x = np.sin(t)-t*np.cos(t)-3.14
my_y = np.cos(t)+t*np.sin(t)+1

plt.plot(x,y, color = 'red')
plt.plot(your_x,your_y, color = 'green')
plt.plot(my_x,my_y, '--', color = "yellow")
```
